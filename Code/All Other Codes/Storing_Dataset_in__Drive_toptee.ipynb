```python
import os
import requests

# Define the directory where you want to save the images
save_dir = "/content/drive/MyDrive/toptee"

# Create the directory if it doesn't exist
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Function to download image from URL and save with provided name
def download_image(url, filename):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open(filename, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded {filename}")
        else:
            print(f"Failed to download {filename}. Status code: {response.status_code}")
    except Exception as e:
        print(f"Exception occurred while downloading {filename}: {e}")

# Function to read code-URL pairs from a file
# Function to read code-URL pairs from a file
def read_code_url_pairs(file_path):
    code_url_pairs = {}
    with open(file_path, 'r') as file:
        for line in file:
            # Split the line by whitespace
            parts = line.strip().split(' ')
            # Extract the code (first part) and join the rest to form the URL
            code = parts[0]
            url = ' '.join(parts[1:])
            code_url_pairs[code] = url
    return code_url_pairs


# Path to the file containing code-URL pairs
file_path = "/content/drive/MyDrive/asin2url.toptee.txt"

# Read code-URL pairs from the file
image_links = read_code_url_pairs(file_path)

# Iterate over each code and download the corresponding image
for code, url in image_links.items():
    filename = os.path.join(save_dir, f"{code}.jpg")
    download_image(url, filename)
```
